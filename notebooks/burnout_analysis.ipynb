```python
# Nan ستبعاد القيم غير المطابقة في عمود الأقسام وتحويلها إلى
valid_departments = ['Engineering - CS / CSE / CSC / Similar to CS', 'Engineering - EEE/ ECE / Similar to EEE', 'Biological Sciences', 'Business and Entrepreneurship Studies', 'Engineering - Civil Engineering / Similar to CE', 'Environmental and Life Sciences', 'Engineering - Mechanical Engineering / Similar to ME' ,'Other']  # list of valid departments
df.loc[~df['4. Department'].isin(valid_departments), '4. Department'] = np.nan
df
```

```python
def map_cgpa_range(x):
    if x == 'Below 2.50':
        return 2.49
    elif x == '2.50 - 2.99':
        return (2.50 + 2.99) / 2
    elif x == '3.00 - 3.39':
        return (3.00 + 3.39) / 2
    elif x == '3.40 - 3.79':
        return (3.40 + 3.79) / 2
    elif x == '3.80 - 4.00':
        return (3.80 + 4.00) / 2
    elif x == 'Other':
        return np.nan
    else:
        return np.nan
df.loc[:, '6. Current CGPA Numeric'] = df['6. Current CGPA'].apply(map_cgpa_range)
```

```python
# إنشاء عمود جديد: هل يعاني الطالب من قلق مرتفع؟ (نعم/لا)
df['High_Anxiety_Flag'] = df['Anxiety Value'].apply(lambda x: 'Yes' if x >= 15 else 'No')
```

```python
years = ['First Year or Equivalent', 'Second Year or Equivalent', 'Third Year or Equivalent', 'Fourth Year or Equivalent']

question1_means = {}

for year in years:
    mean_value = df[df['5. Academic Year'] == year]['1. In a semester, how often you felt nervous, anxious or on edge due to academic pressure? '].mean()
    question1_means[year] = mean_value

for year, value in question1_means.items():
    print(f"متوسط القلق في السنة {year}: {round(value, 2)}")
```

## 📝 المهمة 4: التقرير النهائي

قمنا بمقارنات لتحليل بيانات القلق الأكاديمي بين طلاب الجامعة استنادًا إلى متغيرات مثل الجنس، السنة الدراسية، والمعدل التراكمي، باستخدام بيانات استبيانية وحساب متوسطات القلق لمختلف الفئات.

أظهرت النتائج أن متوسط القلق لدى الإناث (حوالي 14.57) أعلى من الذكور (حوالي 11.86)، ما يشير إلى وجود فروق في التجربة الأكاديمية والضغوط النفسية بين الجنسين.

كما استخدمنا عمليات منطقية AND وOR لتحليل القلق بناءً على السنة الدراسية والمعدل التراكمي:


باستخدام AND، وجدنا أن طلاب السنة الرابعة ممن يقل معدلهم عن 3 يعانون من أعلى مستوى قلق (15.0)، ما قد يعكس ضغوط التخرج أو تدني الأداء.


باستخدام OR، شملت المقارنة طلاب السنة الأولى أو من لديهم معدل 3 فأكثر، وبلغ متوسط القلق لديهم حوالي 12.60، وهو أقل، مما قد يشير إلى تفاؤل أو استقرار نفسي نسبي في هذه الفئة.


وفي مقارنة متوسط القلق حسب السنة الدراسية، تبين أن السنة الثانية سجلت أعلى مستوى قلق (2.14)، بينما كانت السنة الأولى الأقل (1.44)، ما يعكس تزايد القلق بعد الدخول للجامعة، ثم انخفاضه تدريجيًا مع التكيف في السنوات اللاحقة.

```python
import matplotlib.pyplot as plt
import seaborn as sns

# إعداد الشكل
plt.figure(figsize=(7, 7))

# حساب نسب مستويات القلق
anxiety_counts = df['Anxiety Label'].value_counts()

# رسم pie chart
colors = sns.color_palette('pastel')[0:5]
plt.pie(anxiety_counts, labels=anxiety_counts.index, colors=colors, autopct='%.1f%%', startangle=140)

# العنوان
plt.title('Distribution of Anxiety Levels Among Students')
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 6))

# حساب المتوسط لكل سنة دراسية
academic_mean = df.groupby('5. Academic Year')['Anxiety Value'].mean().sort_values()

# رسم barplot
sns.barplot(x=academic_mean.index, y=academic_mean.values, palette='viridis')

# تدوير أسماء السنوات لسهولة القراءة
plt.xticks(rotation=45, ha='left')

# إضافة العنوان والمحاور
plt.xlabel('Academic Year')
plt.ylabel('Average Anxiety Value')
plt.title('Average Anxiety Value by Academic Year')

plt.tight_layout()
plt.show()
```

```python
# رسم countplot
sns.countplot(data=df, x='High_Anxiety_Flag', palette='Set2')

# إضافة العنوان والمحاور
plt.title('Number of Students with High Anxiety')
plt.xlabel('High Anxiety')
plt.ylabel('Number of Students')

plt.show()
```

### قسم الاستنتاجات والتقرير التفسيري

أظهرت نتائج التحليل الإحصائي أن القلق الأكاديمي منتشر بشكل واضح بين طلاب الجامعة، مع وجود فروق ملحوظة حسب الجنس والسنة الدراسية والمعدل التراكمي. وتشير البيانات إلى أن الإناث يعانين من مستويات قلق أعلى مقارنة بالذكور، مما قد يعكس اختلافات في الضغوط النفسية أو توقعات الأداء.

• الجنس: كانت مستويات القلق أعلى لدى الطالبات مقارنة بالطلاب.
 قد يكون السبب في ذلك اختلاف الضغوط النفسية أو التوقعات الدراسية
• السنة الدراسية: طلاب السنة الثانية أظهروا أعلى درجات القلق، ويقل تدريجيًا بعد ذلك.
 هذا يوحي بأن مرحلة التكيف مع الجامعة ومتطلباتها قد تكون الأصعب
• المعدل التراكمي: طلاب السنة الرابعة الذين تقل معدلاتهم عن 3.0 يعانون من أعلى درجات  
 القلق غالبًا ، بسبب ضغوط التخرج والخوف من تدني الأداء
• مستوى القلق بشكل عام: ما يقارب 40٪ من الطلاب يعانون من قلق شديد، والنسبة الأقل لديهم قلق بسيط أو معدوم.
مما يدل على أن القلق منتشر بين الطلاب، لكنه لا يشمل الجميع.
• البيانات العددية: تم رصد 41 طالبًا يعانون من قلق مرتفع، مقابل 60 لا يعانون من قلق حاد.
وهذا يؤكد وجود القلق بشكل واضح، لكنه ليس عامًا على كل الطلبة
